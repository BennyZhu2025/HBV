```python
#01original data extracting
import warnings
import pandas as pd,numpy as np

# Suppress warnings
warnings.filterwarnings("ignore")

# Define the file path
file_path = "2015-2022数据汇总表(有两个时间点！！).xlsx"

# Load the data
df = pd.read_excel(file_path)

# Define the new column names
column_mapping = {
    "ID":"ID",
    "年龄": "Age",
    "户籍": "Hukou",
    #"既往乙肝母婴阻断史": "History of PMTCT",
    "孕次": "Gravida",
    "分娩时产次": "Parity",
    "身高": "Height",
    "分娩BMI": "BMI at delivery",
    "孕前BMI": "Pre-pregnancy BMI",
    "孕周": "Gestational week",
    "胎数": "Number of infants",
    "分娩方式": "Cesarean delivery",
    "新生儿性别": "Newborn gender",
    #"体重": "Newborn weight",
    #"身高": "Newborn height",
    #"头围": "Newborn head circumference",
    "妊娠期AVT史": "AVT during pregnancy",
    "AVT总时长": "Duration of AVT during pregnancy",
    #"用药前（或孕28周前）孕周": "Gestational age before AVT",
    "基线ALT1": "ALT",
    "基线AST1": "AST",
    "基线ALB1": "ALB",
    "基线UREA1": "Urea",
    "基线CR1": "Cr",
    "基线TBIL1": "TBIL",
    "基线DBIL1": "DBIL",
    "基线HGB1": "HGB",
    "基线PLT1": "PLT",
    "基线PT": "PT",
    "基线INR1": "INR",
    "基线APRI1": "APRI",
    "基线FIB-4": "FIB-4",
    #"分娩ALT3": "ALT at delivery",
    #"分娩AST3": "AST at delivery",
    #"分娩ALB3": "ALB at delivery",
    #"分娩TBIL3": "TBIL at delivery",
    #"分娩DBIL3": "DBIL at delivery",
    #"分娩UREA2": "Urea at delivery",
    #"分娩CR2": "Cr at delivery",
    #"分娩HGB2": "HGB at delivery",
    #"分娩PLT2": "PLT at delivery",
    #"分娩PT": "PT at delivery",
    #"分娩INR2": "INR at delivery",
    #"分娩APRI": "APRI at delivery",
    #"分娩FIB-4": "FIB-4 at delivery",
    "甲状腺疾病具体": "Thyroid disease",
    "地中海贫血": "Thalassemia",
    "G6PD缺乏": "G6PD deficiency",
    "前置胎盘伴或不伴出血": "Placenta previa",
    "妊娠合并肥胖症": "Obesity",
    "妊娠糖尿病": "Gestational diabetes",
    "妊娠期高血压疾病": "Gestational hypertension",
    "子痫前期": "Preeclampsia",
    "妊娠合并ICP": "ICP", # intrahepatic cholestasis of pregnancy
    #"胎盘形态异常": "Abnormal placental morphology",
    #"是否合并羊水量异常": "Abnormal amniotic fluid volume",
    #"超声脂肪性变": "Ultrasound fatty change",
    #"IVF-ET术后": "IVF-ET",
    "日期1": "Baseline date",
    "分娩日期": "Delivery date",
    "肝炎分期分组": "Stage of hepatitis",
    #"孕晚-分娩HBV DNA3": "HBV DNA at delivery",
    "基线HBsAg1": "HBsAg at baseline",
    "基线HBeAg1": "HBeAg at baseline",
    "基线HBV DNA1": "HBV DNA at baseline",
    #"1分钟阿氏评分": "1-minute Apgar score",
    "低体重儿": "Low birth weight infant",
    "产后出血": "Postpartum hemorrhage",
    "早产": "Preterm birth",
    "胎膜早破": "PROM",  # premature rupture of membranes
    "宫内窘迫": "Intrauterine distress",
    "新生儿窒息": "Neonatal asphyxia",
    "是否合并出生缺陷": "Birth defects",
    "是否有死婴": "Stillbirth",
    "小于胎龄儿":"SGA",
    "大于胎龄儿":"LGA"
}

# Rename columns based on the mapping
df.rename(columns=column_mapping, inplace=True)

# Filter the DataFrame to retain only the renamed columns
columns_to_keep = list(column_mapping.values())
df = df[columns_to_keep]

# Replace specified values
binary_columns = [
    "Hukou",
    "Gestational diabetes",
    "Gestational hypertension",
    "ICP",
    "Thalassemia",
    "G6PD deficiency",
    "Placenta previa",
    #"Abnormal placental morphology",
    "Obesity",
    "Low birth weight infant",
    "Postpartum hemorrhage",
    "Preterm birth",
    "PROM",
    "Intrauterine distress",
    "Neonatal asphyxia",
    "Birth defects",
    "Stillbirth",
    "SGA",
    "LGA",
    "Preeclampsia",
    #"IVF-ET",
    #"Abnormal amniotic fluid volume"
]

df[binary_columns] = df[binary_columns].replace(2, 0)
df["Thyroid disease"] = df["Thyroid disease"].replace(3, 0)
df["Cesarean delivery"] = df["Cesarean delivery"].replace("2.剖宫产", 1).replace(to_replace=r'.*', value=0, regex=True)
df["Newborn gender"] = df["Newborn gender"].replace("1男", 1).replace(to_replace=r'.*', value=0, regex=True)
df["AVT during pregnancy"] = df["AVT during pregnancy"].replace("4无", 0).replace(to_replace=r'.*', value=1, regex=True)
df["AVT during pregnancy"] = df["AVT during pregnancy"].apply(lambda x: 1 if x != 0 else 0)
#df["Ultrasound fatty change"] = df["Ultrasound fatty change"].replace("无", 0).replace(to_replace=r'.*', value=1, regex=True)
df["Birth defects"]=df["Birth defects"].replace("否",0).replace(to_replace=r'.*', value=1, regex=True)

# Create binary columns for hyperthyroidism and hypothyroidism
df["Hyperthyroidism"] = (df["Thyroid disease"] == 2).astype(int)
df["Hypothyroidism"] = (df["Thyroid disease"] == 1).astype(int)

# Ensure dates are in datetime format
#df["Baseline date"] = pd.to_datetime(df["Baseline date"], format='%m/%d/%Y')
#df["Delivery date"] = pd.to_datetime(df["Delivery date"], format='%m/%d/%Y %H:%M')

# Calculate duration in days
#df["Duration"] = (df["Delivery date"] - df["Baseline date"]).dt.days

# Initialize the 'Stage' column
df["Stage"] = df["Stage of hepatitis"]

# Update the 'Stage' column based on conditions
'''
df.loc[
    (df["Stage of hepatitis"] == 5) & 
    ((df["HBeAg at baseline"].isnull()) | (df["HBeAg at baseline"] < 1)) & 
    (df["HBV DNA at baseline"] >= 2000) & 
    (df["ALT"] <= 25), 
    "Stage"] = 6

df.loc[
    (df["Stage of hepatitis"] == 5) & 
    ((df["HBeAg at baseline"].isnull()) | (df["HBeAg at baseline"] < 1)) & 
    (df["HBV DNA at baseline"] >= 2000) & 
    (df["ALT"] > 25), 
    "Stage"] = 7
'''
# Drop unnecessary columns
df = df.drop(columns=["Delivery date", "Baseline date", "Thyroid disease","Stage of hepatitis"]).replace("未知",None)
df.replace("未知", None, inplace=True)

# Ensure column names are properly quoted and handle fillna
df[['HBsAg at baseline', 'HBeAg at baseline']] = df[['HBsAg at baseline', 'HBeAg at baseline']].fillna(0)
# Apply thresholding to the specified columns
df[['HBsAg at baseline', 'HBeAg at baseline']] = (df[['HBsAg at baseline', 'HBeAg at baseline']] > 10).astype(int)
# Apply log transformation, handling any potential zero or negative values
df['HBV DNA at baseline'] = df['HBV DNA at baseline'].apply(lambda x: np.log(x) if x > 0 else np.nan)

# Check for string values in the DataFrame
string_values = df.iloc[:, 1:].applymap(lambda x: isinstance(x, str))
str_values = df.iloc[:, 1:][string_values].stack().unique()
print("String values in the DataFrame are:", str_values)

# Keep only the first 510 rows
df = df.iloc[:510, :]

# Remove rows with 3 consecutive missing values
rows_to_drop = df.apply(lambda row: row.isnull().rolling(window=3).sum().max() == 3, axis=1)
df = df[~rows_to_drop]

# Reset the index
df.reset_index(drop=True, inplace=True)

# Display the modified DataFrame
df.to_csv('01data_original.csv', index=False, encoding="utf-8-sig")
```
